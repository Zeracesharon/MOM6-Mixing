```python
import numpy as np

def compute_vertical_temp_gradient(T, z):
    """
    Compute the vertical gradient ∂Θ/∂z of your LES temperature field.

    Parameters
    ----------
    T : array_like, shape (ntime, nz)
        LES potential temperature [°C] or [K], time × vertical grid.
    z : array_like, shape (nz,)
        Corresponding depths [m], positive downward.

    Returns
    -------
    dTdz : ndarray, shape (ntime, nz)
        Vertical gradient ∂T/∂z at each time and level.
    """
    # np.gradient handles nonuniform z automatically when you pass it along
    return np.gradient(T, z, axis=1)


def compute_hML_from_gradient(T, z):
    """
    Compute the mixed‐layer depth hML defined by the maximum vertical
    temperature gradient |∂Θ/∂z|.

    Parameters
    ----------
    T : array_like, shape (ntime, nz)
        LES potential temperature [°C] or [K].
    z : array_like, shape (nz,)
        Depth grid [m], positive downward.

    Returns
    -------
    hML : ndarray, shape (ntime,)
        Depth of the maximum |∂T/∂z| at each time step.
    grad  : ndarray, shape (ntime, nz)
        The full vertical gradient arrays ∂T/∂z (for plotting or diagnostics).
    """
    dTdz = compute_vertical_temp_gradient(T, z)     # shape (ntime,nz)
    # find index of max absolute gradient at each time
    idx  = np.argmax(np.abs(dTdz), axis=1)          # shape (ntime,)
    # convert indices to depths
    hML  = z[idx]
    return hML, dTdz


# --- Example usage ---
# T_LES : (ntime,nz) your LES Θ field in °C (or K)
# zl    : (nz,) depth array in m

hML_LES, grad_LES = compute_hML_from_gradient(prog_LES.temp, prog_LES.zl)
hML_LES_noLT, grad_LES_noLT = compute_hML_from_gradient(prog_LES_noLT.temp, prog_LES_noLT.zl)

f,a=plt.subplots(1,1,figsize=(5,3))
a.plot(MOMtime,hML_LES,'b-',label='MOM_1200_1200')
a.plot(MOMtime_noLT,hML_LES_noLT,'b--',label='MOM_1200_1200 noLT')

a.plot(MOMtime,visc_LES.MLD_003,'rx',label='MLD_003 LES')
a.plot(MOMtime_noLT,visc_LES_noLT.MLD_003,'r-',label='MLD_003 LES noLT')
a.plot(MOMtime,visc_LES.MLD_EN1,'kx',label='MLD_EN1 LES')
a.plot(MOMtime_noLT,visc_LES_noLT.MLD_EN1,'k-',label='MLD_EN1 LES noLT')

a.set_ylim(0,250)
a.set_xlim(0,3)
a.set_ylabel(r'maximum |∂Θ/∂z| m',fontsize=12)
a.set_xlabel(r'day',fontsize=12)
a.legend()


f,ax=plt.subplots(2,2,figsize=(15,6))
levels=np.linspace(-5.0E-1,5.0E-1,20)
cmap=cmocean.cm.balance
cmap.set_bad('gray')
a=ax[0,0]
C=a.pcolormesh(MOMtime,visc_LES.zi,visc_LES.Kd_interface.T,shading='nearest',norm = mpl.colors.BoundaryNorm(levels, ncolors=cmap.N, clip=False),cmap=cmap)
a.set_ylim(200,0)
a.set_xlim(0,3)
a.set_ylabel(r'Depth [m]',fontsize=12)
a.set_xlabel(r'day',fontsize=12)
a.set_title(r'$Kd_interface$ LES ',fontsize=12)
plt.colorbar(C,ax=a)
a=ax[0,1]
C=a.pcolormesh(MOMtime,visc_LES_noLT.zi,visc_LES_noLT.Kd_interface.T,shading='nearest',norm = mpl.colors.BoundaryNorm(levels, ncolors=cmap.N, clip=False),cmap=cmap)
a.set_ylim(200,0)
a.set_xlim(0,3)
a.set_ylabel(r'Depth [m]',fontsize=12)
a.set_xlabel(r'day',fontsize=12)
a.set_title(r'$Kd_interface$ LES noLT ',fontsize=12)
plt.colorbar(C,ax=a)

a=ax[1,0]
levels=np.linspace(-5.0E-1,5.0E-1,20)
cmap=cmocean.cm.balance
cmap.set_bad('gray')
C=a.pcolormesh(MOMtime,visc_LES.zi,(visc_LES.Kd_interface-visc_LES_noLT.Kd_interface).T,shading='nearest',norm = mpl.colors.BoundaryNorm(levels, ncolors=cmap.N, clip=False),cmap=cmap)
a.set_ylim(200,0)
a.set_xlim(0,3)
a.set_ylabel(r'Depth [m]',fontsize=12)
a.set_xlabel(r'day',fontsize=12)
a.set_title(r'$Kd_interface$ LES - LES noLT ',fontsize=12)
plt.colorbar(C,ax=a)
a=ax[1,1]
C=a.pcolormesh(MOMtime,visc_LES.zi,(visc_LES.Kd_ePBL-visc_LES_noLT.Kd_ePBL).T,shading='nearest',norm = mpl.colors.BoundaryNorm(levels, ncolors=cmap.N, clip=False),cmap=cmap)
a.set_ylim(200,0)
a.set_xlim(0,3)
a.set_ylabel(r'Depth [m]',fontsize=12)
a.set_xlabel(r'day',fontsize=12)
a.set_title(r'$Kd_ePBL$ LES - LES no LT ',fontsize=12)
plt.colorbar(C,ax=a)
f.tight_layout()

```
